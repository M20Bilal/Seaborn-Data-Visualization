```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from seaborn
titanic = sns.load_dataset("titanic")

# Set up the figure and axes for plotting
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
sns.set_palette("pastel")

# Plot 1: Survival by Passenger Class
sns.countplot(ax=axes[0, 0], data=titanic, x="class", hue="survived")
axes[0, 0].set_title("Survival by Passenger Class")
axes[0, 0].set_xlabel("Passenger Class")
axes[0, 0].set_ylabel("Count")

# Plot 2: Survival by Sex
sns.countplot(ax=axes[0, 1], data=titanic, x="sex", hue="survived")
axes[0, 1].set_title("Survival by Sex")
axes[0, 1].set_xlabel("Sex")
axes[0, 1].set_ylabel("Count")

# Plot 3: Survival by Age Group
titanic["age_group"] = pd.cut(titanic["age"], bins=[0, 18, 30, 50, 80])
sns.countplot(ax=axes[1, 0], data=titanic, x="age_group", hue="survived")
axes[1, 0].set_title("Survival by Age Group")
axes[1, 0].set_xlabel("Age Group")
axes[1, 0].set_ylabel("Count")

# Plot 4: Survival by Embarked Port
sns.countplot(ax=axes[1, 1], data=titanic, x="embarked", hue="survived")
axes[1, 1].set_title("Survival by Embarked Port")
axes[1, 1].set_xlabel("Embarked Port")
axes[1, 1].set_ylabel("Count")

# Adjust the spacing between plots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from seaborn
titanic = sns.load_dataset("titanic")

# Set up the figure and axes for plotting
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
sns.set_palette("pastel")

# Plot 1: Survival by Passenger Class and Sex
sns.barplot(ax=axes[0, 0], data=titanic, x="class", y="survived", hue="sex")
axes[0, 0].set_title("Survival by Passenger Class and Sex")
axes[0, 0].set_xlabel("Passenger Class")
axes[0, 0].set_ylabel("Survival Rate")

# Plot 2: Survival by Embarked Port and Pclass
sns.barplot(ax=axes[0, 1], data=titanic, x="embark_town", y="survived", hue="class")
axes[0, 1].set_title("Survival by Embarked Port and Passenger Class")
axes[0, 1].set_xlabel("Embarked Port")
axes[0, 1].set_ylabel("Survival Rate")

# Plot 3: Survival by Age Group and Sex
titanic["age_group"] = pd.cut(titanic["age"], bins=[0, 18, 30, 50, 80])
sns.barplot(ax=axes[1, 0], data=titanic, x="age_group", y="survived", hue="sex")
axes[1, 0].set_title("Survival by Age Group and Sex")
axes[1, 0].set_xlabel("Age Group")
axes[1, 0].set_ylabel("Survival Rate")

# Plot 4: Survival by SibSp (Number of Siblings/Spouses Aboard)
sns.barplot(ax=axes[1, 1], data=titanic, x="sibsp", y="survived")
axes[1, 1].set_title("Survival by Number of Siblings/Spouses Aboard")
axes[1, 1].set_xlabel("Number of Siblings/Spouses Aboard")
axes[1, 1].set_ylabel("Survival Rate")

# Adjust the spacing between plots
plt.tight_layout()

# Show the plots
plt.show()
```
